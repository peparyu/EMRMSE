<p id="#regression"><b>1.A</b></p>

To study the effects of the incinerator location on housing price, consider the
simple regression model $log(price) = \beta_0 + \beta_1*log(dist) + u$, where **price** is housing price in dollars and **dist** is distance from the house to the incinerator measured in feet. Interpreting this equation causally, what sign do you expect for $\beta_1$ if the presence of the incinerator depresses housing prices? Estimate this equation and interpret the results.

```python
model = smf.ols(formula = "np.log(price)~np.log(dist)", data=df.loc[df['year']==1981])
res = model.fit()
print(res.summary())
```

```python
# Estimate of coefficients are
Intersept = float(res.summary().tables[2].data[1][1])
log_dist = float(res.summary().tables[2].data[2][1])
```

```python
print(f"Increasing distance at 1% increase price at {log_dist}%")
```

2. To the simple regression model in part (i), add the variables *log(intst)*, *log(area)*, *log(land)*, *rooms*, *baths*, and *age*, where *intst* is distance from the home to the interstate, *area* is square *footage* of the house, land is the lot size in square feet, rooms is total number of rooms, baths is number of bathrooms, and age is age of the house in years. Now, what do you conclude about the effects of the incinerator? Explain why (i) and (ii) give conflicting results.

```python
model_ii = smf.ols(formula = "np.log(price)~np.log(dist)+np.log(intst) + np.log(area)+np.log(land) + rooms + baths + age", data=df.loc[df['year']==1981])
res_ii = model_ii.fit()
print(res_ii.summary())
```

<p id="#log2_mod"><b>1.C</b></p>

Add $[log(intst)]^2$ to the model from part (ii). Now what happens? What do you conclude about the importance of functional form?

```python
model_iii = smf.ols(formula = "np.log(price)~np.log(dist)+np.log(intst) + I(np.log(intst) ** 2.0) + np.log(area)+np.log(land) + rooms + baths + age", data=df.loc[df['year']==1981])
res_iii = model_iii.fit()
print(res_iii.summary())
```

<p id="#significant"><b>1.E</b></p>

Is the square of $log(dist)$ significant when you add it to the model from part (iii)?

```python
p_val_log_dist = float(res_iii.summary().tables[1].data[2][4])
p_val_log_dist
```

```python
print(f"log(dist) is significant, because {p_val_log_dist} less {0.05} ")
```

```python
# df['age'] = df['age']-25
df['age_min25'] = (df['age']-25)
df
```

```python
model = smf.ols(formula='nettfa~inc + age + I(age_min25**2)', data =df)
res = model.fit()
print(res.summary())
```

To state the null hypothesis that another year of geneal workforce experitnce has the same effect as another year 
$$
H_0: \theta_2 = 0
$$
$$
H_1: \theta_2 \neq 0
$$

\begin{equation}\label{tstat}
    t = \frac{\hat{\theta}_2}{se(\hat{\theta}_2)}
\end{equation}

```python
float(res.summary().tables[1].data[3][2])
```

```python
# Sample sizea
n = len(df)
# Critical t-statistic with 10% confidence interval
t_crit = st.t.ppf(q=0.95, df=n - 1)
#theta_2
theta_2 = res.params[2]
# se theta_2
se_theta2 = float(res.summary().tables[1].data[3][2])
# t-statistic
t_static = (theta_2)/(se_theta2)
```

```python
if (t_static < t_crit):
    print("Reject hypothesis H1. theta2 = 0")
else:
    print("Reject hypothesis H0. theta2 < 0")
```
