<p id="conside"><b>1.A</b></p>
Consider the standard wage equation $log(wage) =  \beta_0 + 
\beta_1*educ + \beta_2*exper + \beta_3*tenure + u$. State the null hypothesis that another year of general workforce experience has the same effect on $log(wage)$ as another year of tenure with the current employer.

```python

model = smf.ols(formula='np.log(wage)~educ+exper+tenure', data = df)
res = model.fit()
print(res.summary())
```

To state the null hypothesis that another year of geneal workforce experitnce has the same effect as another year 
$$
H_0: \beta_2 = \beta_3
$$
$$
H_1: \beta_2 > \beta_3
$$
To simplify hypotheses, rewrite the null alternative as $H_0: \beta_2 - \beta_3 = 0$ and $H_1: \beta_1 - \beta_2 < 0$ respectively. The $t$ staticstic is based on whether the estimatted difference $\beta_2 - \beta_3$ is sufficiently bigger then zero. That's why 
\begin{equation}\label{tstat}
    t = \frac{\hat{\beta}_2-\hat{\beta}_3}{se(\hat{\beta}_2 - \hat{\beta}_3)}
\end{equation}

```python
res.summary()
```

```python
res.summary().tables[1]
```

```python
res.summary().tables[1].data[3][2]
```

```python
# Sample size
n = len(df)
# Critical t-statistic with 10% confidence interval
t_crit = st.t.ppf(q=0.95, df=n - 1)
#beta2
b2 = res.params[2]
#beta3
b3 = res.params[3]
# se beta2
se_b2 = float(res.summary().tables[1].data[2][2])
# se_beta3
se_b3 = float(res.summary().tables[1].data[3][2])
# t-statistic
t_static = (b2-b3)/(se_b2-se_b3)
```

```python
if (t_static < t_crit):
    print("Reject hypothesis H1. b2 = b3")
else:
    print("Reject hypothesis H0. b2 > b3")
```

<p id="test"><b>1.B</b></p>
Test the null hypothesis in part (i) against a two-sided alternative, at the 5% significance level, by constructing a 95% confidence interval. What do you conclude?

```python
# Sample size
n = len(df)
# Critical t-statistic with 5% confidence interval
t_crit = st.t.ppf(q=0.95, df=n - 1)
#beta2
b2 = res.params[2]
#beta3
b3 = res.params[3]
# se beta2
se_b2 = float(res.summary().tables[1].data[2][2])
# se_beta3
se_b3 = float(res.summary().tables[1].data[3][2])
# t-statistic
t_static = (b2-b3)/(se_b2-se_b3)
```

```python
if (t_static < t_crit):
    print("Reject hypothesis H1. b2 = b3")
else:
    print("Reject hypothesis H0. b2 > b3")
```
